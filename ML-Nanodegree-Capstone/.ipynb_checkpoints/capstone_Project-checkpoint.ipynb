* I need to explore how the sucide rates has evloved through the years so i need to get some some correlation between data to get some insights  


```python
def correlation_plot(data, year):
    df = data[data["year"] == year].drop(columns = "year").copy()
    sns.heatmap(
    df.groupby("country").agg("sum").corr(), cmap = "viridis", annot = True)
    plt.title(f"Correlation among numeric variables in {year}")
    plt.show();
plt.figure(figsize= (5, 3))
correlation_plot(data, 1985)
correlation_plot(data, 2014)
```

### Link between the evolution of suicide and the evolution of other variables

```python
data["year"] = pd.to_datetime(data["year"], format = "%Y")
```

```python
def plot_scatter_links(x:str, y:str, years:list, 
                       agg_fun = "sum", n_col = None):
    
    df = data[data["year"].isin(years)]
    
    df = df.groupby(["year", "country"]).agg(agg_fun).reset_index()
    df["year"] = df["year"].dt.year.astype("object")
    sns.lmplot(x, y, scatter = True, fit_reg= True, 
               data = df, col = "year", col_wrap = n_col, sharex = False, sharey = False)
    
plot_scatter_links(x = "gdp_per_capita ($)", y = "suicides/100k pop", 
                   years = ["1985","1995", "2005", "2015"], n_col= 2, agg_fun= "mean")
plt.subplots_adjust(top=.9)
plt.suptitle("link between  Suicide and GDP/capita", size = 30);
```

```python
plot_scatter_links(x = "population", y = "suicides/100k pop", 
                   years = ["1985","1995", "2005", "2015"], n_col= 2, agg_fun= "mean")
plt.subplots_adjust(top=.9)
plt.suptitle("link between  Suicide and population", size = 30);
```

# Clustering 

I need to see the evlouation of suicide rates from 1985 to 2019 accoriding to gdp_per_capita ($) dropping poplution as i proposed 


I'm adding a new column called occurence which gives tha value 1 to every row so that whenever we group and sum the numeric data we can come back to the original data by dividing it.


```python
data["occurence"] = 1
data["part_generation"] = 0
new_data = data.groupby(["year", "country"]).agg("sum").reset_index()

new_data["year"] = new_data["year"].dt.year
# here i drop the poplution 
new_data.drop(['population'], axis = 1, inplace = True)
print("Wholesale new_Data dataset has {} samples with {} features each.".format(*new_data.shape))
new_data.head()
```

#### Scaling the dataset


```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler()
X = X.fit_transform(new_data.drop(["year", "country"], axis = 1))
selected_columns = ['suicides_no', 'suicides/100k pop', 'HDI for year',
        'gdp_per_capita ($)', 'part_generation','occurence']

new_data[selected_columns] = X
new_data.head()
```

#### Applying Clustering 
Here i'll exmaine the best algorithm to perform clustring 

## K-Means VS GaussianMixture

## 1985 data

```python
from sklearn.cluster import KMeans

# prepraing training data 
x = new_data[new_data["year"] == 1985].drop(["year", "suicides_no"], 
                                                      axis = 1).set_index("country")
```

##### K-Means

```python
from sklearn.metrics import silhouette_score
inertia =[] 
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=0).fit(x)
    preds = kmeans.predict(x)
    inertia.append(silhouette_score(x, preds))
ax = plt.plot(range(2,10),inertia)
```

##### GaussianMixture

```python
from sklearn.mixture import GaussianMixture
inertia =[] 
for k in range(2, 10):
    clusterer = GaussianMixture(n_components = k)
    clusterer.fit(x)
    preds = clusterer.predict(x)
    inertia.append(silhouette_score(x, preds))
    
ax = plt.plot(range(2,10),inertia)
```

### iI k means at 3 clusters gives the best score than GaussianMixture with 7 clusters 

## 2015 data

```python
# prepraing training data 
x_2015 = new_data[new_data["year"] == 2015].drop(["year", "suicides_no"], 
                                                      axis = 1).set_index("country")
```
